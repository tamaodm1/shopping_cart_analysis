```python
import pandas as pd
import time
from mlxtend.frequent_patterns import fpgrowth, association_rules

basket_bool = pd.read_parquet("../data/processed/basket_bool.parquet")
basket_bool.head()
```

```python
start = time.time()

freq_fp = fpgrowth(
    basket_bool,
    min_support=0.01,
    use_colnames=True
)

rules_fp = association_rules(
    freq_fp,
    metric="confidence",
    min_threshold=0.3
)

fp_time = time.time() - start

freq_fp.shape, rules_fp.shape, fp_time
```
